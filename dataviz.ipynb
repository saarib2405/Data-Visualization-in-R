**DATA VISUALIZATION IN R**

*Histogram*

```r
#make this example reproducible
set.seed(1)

# Define data
x1 = rnorm(1000, mean=0.8, sd=0.2)
x2 = rnorm(1000, mean=0.4, sd=0.1)

# Create histograms

hist(x1, col='lavender', xlim=c(0, 1.5), main='Multiple Histograms', xlab='x')
hist(x2, col='maroon', add=TRUE)


# Add legend
legend('topright', legend=c('x1 variable', 'x2 variable'), fill=c('lavender', 'maroon'))
```

*Bar charts using ggplot*

```r
install.packages("ggplot2")
library("ggplot2")
ggplot(mtcars, aes(x =cyl))+geom_bar()
```

*All plots*

```r
#All-plots
plot(mtcars$hp, mtcars$mpg, xlab = "Horsepower", ylab = "Miles per Gallon", main = "Scatter Plot: MPG vs. HP")

pairs(iris[, 1:4], col = iris$Species, pch = 19, main = "Pairwise Scatter Plot: Iris Data")

boxplot(mpg ~ am, data = mtcars, xlab = "Transmission", ylab = "Miles per Gallon", main = "Box Plot: MPG by Transmission Type")

barplot(table(mtcars$cyl), xlab = "Cylinders", ylab = "Frequency", main = "Bar Plot: Distribution of Cylinders")

boxplot(mpg ~ cyl, data = mtcars, xlab = "Cylinders", ylab = "Miles per Gallon", main = "Box Plot: MPG by Cylinder Count")

pairs(mtcars[, c("mpg", "hp", "wt", "qsec")], main = "Pairwise Scatter Plot: Key Variables")

hist(iris$Sepal.Width,col='yellow',main='hisr',xlab='mm',breaks = 10)

species_count<-table(iris$Species)
barplot(species_count,
        main="COUNT OF IRIS SPECIES",
        xlab="SPECIES",
        ylab="COUNT",
        col='white')
ggplot(iris, aes(Species,Sepal.Length)) + geom_boxplot()
boxplot(iris,aes(Sepal.Length,Sepal.Width)) + geom_boxplot()
```
